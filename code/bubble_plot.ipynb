## Clean

```python
merged_df.replace("..", np.nan, inplace=True) #replace ".." to na
merged_df = merged_df.dropna(subset=['Population, total [SP.POP.TOTL]']) #drop all rows with na in population column 
merged_df = merged_df.dropna(subset=['Log GDP per capita']) #drop all rows with na in Log GDP per capita column 
merged_df['Population, total [SP.POP.TOTL]'] = merged_df['Population, total [SP.POP.TOTL]'].astype(float) #change type
```

```python
merged_df = merged_df[~merged_df['year'].isin([2005, 2006, 2007])] #did not use 2005-2007 due to less data
```

```python
merged_df.columns # all column names
```

```python
#change names
merged_df = merged_df.rename(columns={'Life Ladder': 'Happiness Score'})
merged_df = merged_df.rename(columns={'Healthy life expectancy at birth': 'Health'})
merged_df = merged_df.rename(columns={'Perceptions of corruption': 'Corruption'})
merged_df = merged_df.rename(columns={'CO2 emissions (kt) [EN.ATM.CO2E.KT]': 'CO2 emissions'})
merged_df = merged_df.rename(columns={'Population, total [SP.POP.TOTL]': 'Population'})
merged_df = merged_df.rename(columns={'Unemployment, total (% of total labor force) (national estimate) [SL.UEM.TOTL.NE.ZS]': 'Unemployment'})
merged_df = merged_df.rename(columns={'Children out of school, primary [SE.PRM.UNER]': 'Children out of school'})
merged_df = merged_df.rename(columns={'Adjusted net national income (current US$) [NY.ADJ.NNTY.CD]': 'Adjusted net national income'})
merged_df["continent"] = merged_df["continent"].replace("Australia", "Oceania")
merged_df.columns
```

## Save the file

```python
merged_df.to_csv('../data/combine_data.csv') #save file
merged_df
```

## Bubble Plot

```python
# bubble plot with x is log gpt per capita and y is happiness score. It has a time slider from 2008-2021. Group by continent and the size of the bubble is pupulation size.
fig = px.scatter(merged_df, x="Log GDP per capita", y="Happiness Score", animation_frame="year", animation_group="Country",
           size="Population", color="continent", hover_name="Country", size_max=70, color_discrete_map={
               "Oceania": "rgb(166, 216, 84)", #set color
               "North America": "rgb(231, 138, 195)",
               "Europe": "rgb(141, 160, 203)",
               "South America": "rgb(255, 217, 47)",
               "Asia": "rgb(252, 141, 98)",
               "Africa": "rgb(102, 194, 165)",
           }, category_orders={ #set order
               "continent": ["Oceania", "North America", "Europe", "South America", "Asia", "Africa"]
           })

fig.update_layout(title="GDP Per Capita vs Happiness Score ",
                  xaxis_title="Log GDP Per Capita (Dollar)",
                  yaxis_title="Happiness Score (range 0 to 10)",
                  template="plotly_white",
                  legend_title_text='Continent')#update the title and background


fig.show()
```
